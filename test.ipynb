```python
import numpy as np
import string
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def id_generator(size=6, chars=string.ascii_uppercase + string.digits):
    return ''.join(random.choice(chars) for _ in range(size))
```

```python
df = pd.DataFrame(columns=['machine'])
df
```

```python
for i in range(1,96):
    df = df.append({'machine': id_generator()}, ignore_index=True)
```

```python
df
```

```python
WIP = [random.randint(0, 10000) for _ in range(95)]
MOVE = [random.randint(0, 10000) for _ in range(95)]
```

```python
df['WIP'] = WIP
df['MOVE'] = MOVE
```

```python
df
```

```python
quotient, remainder = divmod(len(df), 49)
```

```python
if remainder > 0 :
    pic_number = quotient + 1
```

```python
df.iloc[0]['WIP']
```

```python
for picNo in range(1,pic_number):
    pic_count = 0
    fig, axs = plt.subplots(7, 7)
    for row in range(0,6):
        for col in range(0,6):
            axs[row,col].scatter(df.iloc[col]['WIP'],df.iloc[col]['MOVE'])
            pic_count = pic_count+1
            if pic_count == 49:
                plt.show()
                picNo = picNo +1
```
